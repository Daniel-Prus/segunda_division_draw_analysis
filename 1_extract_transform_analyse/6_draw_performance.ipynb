```python
import numpy as np
import pandas as pd
from draw_analysis.draw_analysis import DrawAnalysis
from draw_analysis.team_draw_analysis import TeamDrawAnalysis
from draw_analysis.visuals import draw_performance_visual
```

```python
fixtures = pd.read_csv('backup/segunda_division_fixtures_2018_2019_2020_2021_07_13.csv')
```

```python
segunda_divison = TeamDrawAnalysis(fixtures)
```

```python
draw_performance = segunda_divison.draw_results_performance()
draw_performance.head()
```

```python
draw_performance_visual(draw_performance, title="Draw performance seasons - 2018, 2019, 2020")
```

```python
filter_2018 = draw_performance['draws_pcts_2018(%)'] > 40
filter_2019 = draw_performance['draws_pcts_2019(%)'] > 40
filter_2020 = draw_performance['draws_pcts_2020(%)'] > 40

# teams with the highest draw performance by seasons
filter_teams = draw_performance[filter_2018 | filter_2019 | filter_2020].index.to_list()
```

```python
highest_draw_performance = segunda_divison.draw_results_performance(teams=filter_teams, transpose=True)
highest_draw_performance = highest_draw_performance[highest_draw_performance['draws_pcts'] > 40]
```

```python
draw_performance_visual(highest_draw_performance)
```
